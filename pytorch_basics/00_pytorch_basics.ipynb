### Matrix multiplication 

Two main ways of performing multiplication in neural networks and deep learning:
1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work
2. The resulting matrix has the shape of the **outher dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
# element-wise multiplication
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication-dot
torch.matmul(tensor, tensor)
```

One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
# torch.mm(tensor_A, tensor_B) # torch.mm is the same as torch.matmul (it's an alias for writing less code)
torch.matmul(tensor_A, tensor_B.T)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix tensor shape issues, we can manipulate the shape of one of tensors using a transpose. A transpose switches the axes or dimensions of a given tensor.

```python
tensor_B.T, tensor_B.T.shape
```

```python
# The matrix multiplication operaton works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (smae shape as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

### Reshaping, stacking, squeezing and unsqueezing tensors
* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side
* Squeeze - removes all `l` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute- return a view of the input with dimensions permuted (swapped) in a certain way


```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
# adding an extra dimensions
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# changing the view
z = x.view(1, 9)
z, z.shape
```

```python
# changing z changes x because a view of a tensor shares the same memory as the original input
z[:, 0] = 5
z, x
```

```python
# stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0) # dim=1
x_stacked
```

```python
# func. squeeze() removes all single dim.s from a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# func. unsqueeze() adds a single dim. to a target tensor at a specific dim.
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# add an extra dimension
x_unsqueezed0 = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed0}")
print(f"New shape: {x_unsqueezed0.shape}")
x_unsqueezed1 = x_squeezed.unsqueeze(dim=1)
print(f"\nNew tensor: {x_unsqueezed1}")
print(f"New shape: {x_unsqueezed1.shape}")
```

```python
# rearranges the dims of a target tensor in a spesific order
x_org = torch.rand(size=(112, 112, 3)) # [height, width, color_channels]

# permute the original tensor to rearrange the axis order
x_permuted = x_org.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_org.shape}")
print(f"New shape: {x_permuted.shape}") # [color_channels, height, width]
```

```python
# they share same memory
x_org[0, 0, 0] = 9110
x_org[0, 0, 0], x_permuted[0, 0, 0]
```
